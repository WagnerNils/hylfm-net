```python
import pandas
import numpy

from typing_extensions import Literal
from pathlib import Path
```

```python
df_path = Path("/g/kreshuk/beuttenm/pycharm_tmp/repos/hylfm-net/logs/celestial-sweep-13/test/output_tensors/metrics.h5")
```

```python
def load_and_explode_metrics_df(df_path, along: Literal[None,"z", "y", "x"] = None):
    store = pandas.HDFStore(df_path)
    df = store["df"]
    drop_suffixes = ["-" + suffix for suffix in ["z", "y", "x"] if suffix != along]
    cols_to_drop = [c for c in df.columns if any(c.endswith(suffix) for suffix in drop_suffixes)]
    df = df.drop(cols_to_drop, axis="columns")
    df = df.reset_index().rename(columns = {'index':'it'})
    df = df.set_index(['it']).apply(lambda x: x.explode()).reset_index()
    cols_wo_suffix = [c for c in df.columns if not c.endswith("-" + along)]
    df = df.set_index(cols_wo_suffix)
    a_col_name = df.columns[0]
    print(f'infering length of values along {along} from: {a_col_name}')
    df[along] = df[a_col_name].apply(lambda x: numpy.arange(len(x)))
    df = df.apply(lambda x: x.explode()).reset_index()
    return df
    
load_and_explode_metrics_df(df_path, along="z")
```
